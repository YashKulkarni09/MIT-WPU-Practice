**Matplotlib Basics**

**Pyplot**

1. Draw a line in a diagram from position (0,0) to position (5,100).

```python
import matplotlib.pyplot as plt
x = [0, 5]
y = [0, 100]
plt.plot(x, y)
plt.show()
```

2. Draw a line in a diagram from position (1, 2) to position (5, 6). (With and
without line)


```python
import matplotlib.pyplot as plt
# Define the coordinates
x = [1, 5]
y = [2, 6]
# Plot with line
plt.figure(figsize=(5, 5))
plt.plot(x, y, marker='o', color='black')
plt.title('With Line')
plt.xlim(0, 6)
plt.ylim(0, 7)
plt.grid(True)
plt.show()
# Plot without line
plt.figure(figsize=(5, 5))
plt.scatter(x, y, marker='o', color='black')
plt.title('Without Line')
plt.xlim(0, 6)
plt.ylim(0, 7)
plt.grid(True)
plt.show()
```

3. Draw a line in a diagram from position (1, 5) to (3, 6) then to (9, 15) and
finally to position (10, 20).


```python
import matplotlib.pyplot as plt
x = [1, 3, 9, 10]
y = [5, 6, 15, 20]
plt.plot(x, y, marker='o', color='yellow')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Connecting Points')
plt.grid(True)
plt.show()
```

4. Draw a line in a diagram without X points.

```python
import matplotlib.pyplot as plt
x = [1, 3, 9, 10]
y = [5, 6, 15, 20]
plt.plot(x, y, color='green', linestyle='-')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Without X Points')
plt.grid(True)
plt.show()
```
